```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = np.random.rand(10, 10)

# Create a heatmap with seaborn
sns.heatmap(df)

# Show the plot
plt.show()
```

```python
# Create a heatmap
sns.heatmap(df, xticklabels='Sector Name', yticklabels='Electricity', cbar_kws={'label': 'GHG Emissions KG'})

# Add a title
plt.title('Heatmap of GHG Emissions by Floor Area and Hours per Week')

# Adjust the color map
plt.set_cmap('GnBu')

# Format the axis labels
plt.xlabel('Sector Name')
plt.ylabel('Electricity')

# Show the plot
plt.show()
```

```python
# Create a heatmap
sns.heatmap(df, xticklabels='Total Floor Area', yticklabels='Average Hours per week', cbar_kws={'label': 'GHG Emissions KG'})

# Add a title
plt.title('Heatmap of GHG Emissions by Floor Area and Hours per Week')

# Adjust the color map
plt.set_cmap('Blues')

# Format the axis labels
plt.xlabel('Total Floor Area (sqft)')
plt.ylabel('Average Hours per Week')

# Show the plot
plt.show()
```
