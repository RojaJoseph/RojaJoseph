```python
bar_plot = az.plot.bar(x='Units Sold',y=['Total Revenue','Total Cost','Total Profit'],color=['purple','red','blue'],stacked=True,rot=True)
plt.xticks(rotation=90)
plt.locator_params(nbins=38)
plt.tick_params(axis='y', which='major', labelsize=7)
```

```python
az.plot.barh(x='Item Type',y='Total Revenue',color='blue')
plt.locator_params(nbins=28)
plt.xlabel('Total Revenue')
```
