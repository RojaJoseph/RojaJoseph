```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\LENOVO\Desktop\World Population.csv")
df.head()
```

```python
df = df.drop(columns='#', errors='ignore')
```

```python
df = df.sort_values(by='Population (2023)', ascending=False)
```

```python
top_20_countries = df.nlargest(20, 'Population (2023)')


sns.set(style="whitegrid")

plt.figure(figsize=(14, 8))
sns.barplot(data=top_20_countries, x="Country (or dependency)", y="Population (2023)", palette="muted")
plt.xticks(rotation=45)
plt.title("Population Distribution of Top 20 Countries in 2023")
plt.xlabel("Country (or dependency)")
plt.ylabel("Population (2023)")
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
df['Population (2023)'].plot(kind='line', figsize=(8, 4), title='Population (2023)')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
plt.figure(figsize=(20, 8))


plt.plot(df['World Share'], df['Population (2023)'], marker='o', linestyle='-', color='b')


plt.title('World Share vs. Population in 2023')
plt.xlabel('Population (2023)')
plt.ylabel('World Share')


plt.grid(True)
plt.show()
```

```python
sns.pairplot(df[['Population (2023)', 'Density  (P/Km²)', 'Fert. Rate', 'Med. Age']])
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.distplot(a=df["Migrants (net)"])
plt.show()
```

```python
df[['Land Area  (Km²)','World Share','Urban Pop %']].plot(kind="area",title="Area distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Population (2023)'], fill=True, color='skyblue', label='Population Distribution')
plt.title('Population Distribution with KDE')
plt.xlabel('Population (2023)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.swarmplot(data=df, x=pd.cut(df['Density  (P/Km²)'], bins=5), y='Population (2023)', color='black', alpha=0.5)
sns.violinplot(data=df, x=pd.cut(df['Density  (P/Km²)'], bins=5), y='Population (2023)', palette='pastel')
plt.title('Population Distribution by Population Density Range')
plt.xlabel('Population Density Range')
plt.ylabel('Population (2023)')
plt.show()
```

```python
sns.jointplot(data=df, x='Population (2023)', y='Density  (P/Km²)', kind='reg', marginal_kws=dict(bins=25, fill=False))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.pointplot(data=df, x='Net Change', y='Yearly Change', hue='Urban Pop %', dodge=True, capsize=0.2)
plt.title('Net Change vs. Yearly Change by Urban Population Percentage')
plt.show()
```
