```python
def handle_outliers(df, columns):
    for col in columns:
        # Calculate the IQR for the column
        Q1 = df_1[col].quantile(0.25)
        Q3 = df_1[col].quantile(0.75)
        IQR = Q3 - Q1

        # Define the upper and lower bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Replace values outside the bounds with lower and upper bound
        df_1[col] = np.where(df[col] < lower_bound, lower_bound, df_1[col])
        df_1[col] = np.where(df[col] > upper_bound, upper_bound, df_1[col])

    return df

# Specify the columns to handle outliers
columns_to_handle_outliers = ['property_size', 'property_age', 'bathroom', 'cup_board', 'balconies']

# Apply the function to handle outliers
df_1_handled_outliers = handle_outliers(df_1, columns_to_handle_outliers)

# Display the shape before and after handling outliers
print(f"Shape before handling outliers: {df.shape}")
print(f"Shape after handling outliers: {df_1_handled_outliers.shape}")

# Summary statistics after handling outliers
summary_stats_handled_outliers = df_1_handled_outliers.describe(percentiles=[0.01, 0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.99])
print(summary_stats_handled_outliers)
```

```python
x=df_1.iloc[:,:21].values
y=df_1.iloc[:,-1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
x_test
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)
```

```python
### KNN -NEIGHBOURS
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
